```python
# Visualizing target variable Clicked on Ad
plt.figure(figsize = (14, 6)) 
plt.subplot(1,2,1)            
sns.countplot(x = 'click', data = df)
plt.subplot(1,2,2)
sns.distplot(df["click"], bins = 20)
plt.show()
```

```python
sns.jointplot(x = "C14", y= "C17", data = df) 

''' Do hi attribute jo encoded hai unmme mujhe koi solid relationship dikh raha hai baki sab scattered hai  '''
```

```python
sns.pairplot(df, hue = 'click', vars = ['banner_pos' ,'C14' ,'C17' ], palette = 'husl')


'''
    This is to show cross product of relationships of different attributes wrt to each other 
    Saaro ka run nahi ho raha mere isme too slow ( Yeh karna hai ) 

'''
```

```python
plots =  ['banner_pos','C1','device_type', 'device_conn_type', 'C14',
       'C15', 'C16', 'C17', 'C18', 'C19', 'C20', 'C21' ]
'''
Uni-modal and Bi-modal peaks ke baareme padhna hai of these numerical attributes 
'''

for i in plots:
    plt.figure(figsize = (14, 6))
    plt.subplot(1,2,1)
    sns.boxplot(df[i])
    plt.subplot(1,2,2)
    sns.distplot(df[i],bins= 20)    
    plt.title(i)    
    plt.show()
```

```python
fig = plt.figure(figsize = (12,10))
sns.heatmap(df.corr(), cmap='Blues', annot = True) 
```

Sabse zyada correlation (+ve) C16 aur clicks mein dikh raha tha toh yeh Line char aur Grouped Bar plot mein dekha thoda 

```python
f,ax=plt.subplots(1,2,figsize=(14,5))
df['C16'][df['click']==1].value_counts().sort_index().plot(ax=ax[0])
ax[0].set_title('C16 Vs Clicks')
ax[0].set_ylabel('Count of Clicks')
pd.crosstab(df["click"], df["C16"]).T.plot(kind = 'Bar',ax=ax[1])
#df.groupby(['Month'])['Clicked on Ad'].sum() # alternative code
plt.tight_layout()
plt.show()
```
